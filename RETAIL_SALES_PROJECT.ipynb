```python
import pandas as pd


df = pd.read_csv('Retail_Sales_Data.csv')
df.head()

```

```python
df.tail()
```

```python

df = df.drop_duplicates()

```

```python
df.tail()
```

```python

df['Sales'] = df['Sales'].fillna(df['Sales'].median())
```

```python
df['Ad_Spend'] = df['Ad_Spend'].fillna(0)
```

```python
df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')

```

```python
df['Product_Category'] = df['Product_Category'].fillna('Miscellaneous')
```

```python

df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')

```

```python

df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce').fillna(1).astype(int)

```

```python

df['Returns'] = df['Returns'].astype(str).str.strip().str.lower()
df['Returns'] = df['Returns'].replace({'yes': 'Yes', 'y': 'Yes', 'no': 'No', 'n': 'No', 'nan': 'No'})

```

```python

df.to_csv('Retail_Sales_Cleaned.csv', index=False)
```
